```python
import pandas as pd
import numpy as np
```

#### loading dataset

```python
sales_data=pd.read_csv("Sales_Data_Analysis.csv",encoding="latin8")
sales_data
```

#### handling missing values- cleaning

```python
sales_data.isnull().sum()
```

```python
sales_data=sales_data.dropna(subset=["Description","CustomerID"])
sales_data
```

```python
sales_data.info()
```

#### handling duplicates

```python
sales_data.duplicated().sum()
```

```python
sales_data=sales_data.drop_duplicates()
sales_data
```

```python
sales_data.duplicated().sum()
```

#### calculating basic statistics

```python
## total quantity sold

total_quantity_sold=sales_data["Quantity"].sum()
total_quantity_sold
```

```python
## total revenue

total_revenue=(sales_data["Quantity"]*sales_data["UnitPrice"]).sum()
total_revenue
```

```python
## average unit price

avg_unit_price=sales_data["UnitPrice"].sum()/len(sales_data["UnitPrice"])
avg_unit_price
```

```python
## unique customers

num_of_unique_customers=len(sales_data["CustomerID"].unique())
num_of_unique_customers
```

```python
## total revenue generated by each country

sales_data["Revenue"]=sales_data["Quantity"]*sales_data["UnitPrice"]

sales_data.groupby("Country")["Revenue"].sum()
```

```python
## revenue generated by france

sales_data.groupby("Country")["Revenue"].sum().loc["France"]
```

#### top 10 most sold Products (Description) based on Quantity



```python
product_sales=sales_data.groupby("Description")["Quantity"].sum()
product_sales_sorted=product_sales.sort_values(ascending=False)

print(product_sales_sorted.head(10))
```

#### Extract the month from InvoiceDate and calculate the total sales per month to visualizethe trend and add a new column called “Month” and display the month sales of month “2011-12”.

```python
sales_data["InvoiceDate"]=pd.to_datetime(sales_data["InvoiceDate"])
sales_data
sales_data["Month"]=sales_data["InvoiceDate"].dt.to_period("M")
sales_data

monthly_sales=sales_data.groupby("Month")["Revenue"].sum()
monthly_sales

print(monthly_sales.loc["2011-12"])
```

#### average number of items (Quantity) purchased per invoice for each Customer

```python
number_of_items=sales_data.groupby(["InvoiceNo","CustomerID"])["Quantity"].sum()
avg_items=number_of_items.groupby("CustomerID").mean()
avg_items.loc[12583.0]
```

#### Find total number of countries and identify the top 5 countries with the highest number of unique invoices

```python
print(f"total number of countries: {sales_data["Country"].nunique()}")

invoice_count=sales_data.groupby("Country")["InvoiceNo"].nunique()
invoice_count
top_5=invoice_count.sort_values(ascending=False).head(5)
top_5
```

#### Analyze the distribution of unit prices using NumPy to get the mean, median, and standard deviation.

```python
import numpy as np

unit_prices=sales_data["UnitPrice"].values
print(unit_prices)

print(f"mean price:{np.mean(unit_prices)}")
print(f"median price:{np.median(unit_prices)}")
print(f"standard deviation:{np.std(unit_prices)}")

```

#### Identify transactions where the Quantity is negative (assumed to be refunds) and calculate the total value of refunded products.

```python
negative_quantity=sales_data[sales_data["Quantity"]<0]
negative_quantity

total_negative_value=negative_quantity["Quantity"].sum()
total_negative_value

total_refund_value=(negative_quantity["Quantity"]*negative_quantity["UnitPrice"]).sum()
total_refund_value
```
